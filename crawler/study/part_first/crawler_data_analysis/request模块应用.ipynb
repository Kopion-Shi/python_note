### request模块
- urllib模块：基于模拟浏览器上网的模块，网络请求模块
- request模块：基于网络请求的模块
    - 作用：模拟浏览器上网
    - 编码流程：
        -指定url
        - 发起请求
        - 获取数据（爬取数据）
        - 持久化


```python
# 1.爬取搜狗首页的内容
import requests

url='https://www.sogou.com'
response=requests.get(url)
page_text=response.text#text返回的是字符串形式的响应数据
with open('./sougou.html','w',encoding='utf-8') as f:
    f.write(page_text)
```

```python
#2.网页采集器
#知识点：参数动态化，UA伪装，乱码处理
word=input('enter a key word')
url='https://www.sogou.com/web'
## 参数动态化：将请求的参数封装成字典做那个用到get方法的param参数中
params_dict={
    'query':word
}
response=requests.get(url,params=params_dict)
page_text=response.text
print(page_text)
filename='./'+word+'.html'
with open(filename,'w',encoding='utf-8') as f:
    f.write(page_text)
print('running over!!!!')
```

- 异常的访问请求
    - 在爬虫中正常的访问请求指的是通过真实的浏览器发起的访问请求
    - 异常的访问请求：通过非浏览器发起的请求。（爬虫程序模拟的请求发送)
- 正常的求请访问和异常的请求访问的判别方式是什么？
    - 通过请求头中的User_Agent判别
    - User_Agent：请求载体的身份标识
    - 目前请求的载体可以是：浏览器 爬虫
- 反爬机制：
    - UA检测：网站后台会检测请求载体的身份标识（UA）是不是浏览器
    - 是： 正常的访问请求
    - 不是：异常的访问请求
- 反爬策略：
    - UA伪装：
        - 将爬虫的请求载体身份标识伪装/篡改成浏览器的身份标识
        - user-agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36

```python
# UA 伪装
word=input('enter a key word')
url='https://www.sogou.com/web'
## 参数动态化：将请求的参数封装成字典做那个用到get方法的param参数中
params_dict={
    'query':word
}
#UA伪装
hearder={
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'
}
#将请求头作用在请求的请求头中
response=requests.get(url,params=params_dict,headers=hearder)
page_text=response.text
response.encoding='utf-8' ## 手动修改响应对象的编码格式
print(page_text)
filename='./'+word+'.html'
with open(filename,'w',encoding='utf-8') as f:
    f.write(page_text)
print('running over!!!!')
```

## 需求：爬取豆瓣网的电影详细数据
url：https://movie.douban.com/typerank?type_name=%E7%88%B1%E6%83%85&type=13&interval_id=100:90&action=
- 涉及到的知识点：动态加载数据（目标数据，通过ajax的方式请求返回发到页面）
当前网站分析：
- 当滚轮滑到底部的时候，页面会发起ajax的请求，且请求到一组电影详情数据
- 当滚轮不滑动的时候，页面显示的电影数据，通过浏览器地址栏中的url发起的请求是请求不到的

知识点： 动态加载数据
- 可见非即所得
- 概念：通过非浏览器的url请求到的数据
- 当对一个陌生的数据进行指定数据爬取之前，要校验目标数据是否为动态加载的数据
    - 是
        - 基于抓包工具进行全局搜索，锁定动态加载数据对应的数据包即可，从数据包中提取 请求的url和请求方式和请求参数
    - 否
        - 直接对地址栏的url发起请求即可获取到指定的数据

```python
url='https://movie.douban.com/j/chart/top_list'
params_dict={
    'type': '13',
    'interval_id': '100:90',
    'action': '',
    'start': '0',
    'limit': '100',
}
header={
   ' User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'
}
response=requests.get(url=url,headers=hearder,params=params_dict)
response_page=response.json()#返回序列化好的对象
for dic in response_page:
    name=dic['title']
    score=dic['score']
    print(name,score)
```
