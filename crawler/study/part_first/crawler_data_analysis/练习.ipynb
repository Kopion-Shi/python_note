## 站长素材图片数据的爬取
- 反爬机制：图片懒加载
- 分析：
    - 图片数据是否为动态加载的数据
        - 除了可以在response选项卡中进行局部搜索，可以观察preview这个选项卡中的可视化的内容
        - 发现除了preview中只显示图片的名称，并没有显示图片数据

```python
import requests
import re
url='https://sc.chinaz.com/tupian/meinvxiezhen.html'
header={
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'
}

page_text=requests.get(url=url,headers=header)
page_text.encoding='utf-8'
page_text=page_text.text
print(page_text)
# ex='<div class="item".*?data-original="(.*?)".*?class="lazy".*?alt="(.*?)".*?</div>'
ex='<div class="bot-div".*?href="(.*?)".*?title="(.*?)".*?</div>'
res=re.findall(ex,page_text,re.S)
res
```
